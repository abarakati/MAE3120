### 1  $1^{st}$ order high pass filter

The following RC circuit is a passive first-order high-pass filter.  
<img src="img/HPRCFilter.png" width="240">

(a) Using complex definiation of impedance prove that (show your steps):
\begin{align*}
\frac{V_{out}}{V_{in}} = \frac{i \omega RC}{1 + i \omega RC} 
\end{align*}

One recogonizes a voltage divider:
\begin{align}
V_{out} = V_{in} \frac{Z_R}{Z_R+Z_C}
\end{align}
with $Z_R=R$ and $Z_C = \frac{1}{i\omega C}$
\begin{align*}
\frac{V_{out}}{V_{in}} & = \left( \frac{R}{R + 1/(i \omega C)} \right) \\
 & = \left( \frac{i \omega RC}{1 + i \omega RC} \right) 
\end{align*}

(b) Introduce the cutoff frequency, $f_c$: 
\begin{align*}
f_{c} = \frac{1}{2\pi RC}
\end{align*}
and show that the gain, $G$, is:
\begin{align*}
G = \frac{|V_{out}|}{|V_{in}|} = \frac{1}{\sqrt{1+ \left( \frac{f_{c}}{f}\right)^2 }}
\end{align*}

Substitute $f_{c} = \frac{1}{2\pi RC}$ and remembering that $f = \frac{\omega}{2 \pi}$:

\begin{align*}
\frac{V_{out}}{V_{in}} = \frac{i f/f_c}{1+i f/f_c}
\end{align*}
Divide by $if/f_c$ on top and bottom:
\begin{align*}
\frac{V_{out}}{V_{in}} = \frac{1}{1+i f_c/f}
\end{align*}

The norm of a complex number $z = a+ib$ is $|z|=\sqrt{z\bar{z}}$.  Here 
\begin{align}
z & = \frac{1}{1+if_c/f} = \left( \frac{1}{1+(f_c/f)^2} - \frac{if_c/f}{1+(f_c/f)^2} \right) \\
z\bar{z} & = \left( \frac{1}{1+(f_c/f)^2} \right)^2  \left( 1+ \left(\frac{f_c}{f}\right)^2 \right)
\\
& =  \frac{1}{1+(f_c/f)^2}
\end{align} 
Thus,
\begin{align}
G = \frac{|V_{out}|}{|V_{in}|} =\frac{1}{\sqrt{1+(f_c/f)^2}}
\end{align}




(c) Plot the Bode diagram of this filter:

Note the phase is given as: 
\begin{align*}
\phi = \arctan \left( \frac{f_{c}}{f} \right)
\end{align*}

### 2 Optimum power transfer

<img src="img/OptimalLoad.PNG" width="240">

(a) Show that the power in the load, $P_L$ is:
\begin{align*}
P_L = V_T \left( \frac{R_L}{R_L+R_T} \right)  \left( \frac{V_T}{R_L+R_T} \right)
\end{align*}

Voltage in $R_L$ (voltage divider):
\begin{align}
V_{L} = V_T \frac{R_L}{R_T+R_L}
\end{align}
Current through the circuit:
\begin{align}
I = V_T \frac{1}{R_T+R_L}
\end{align}
Power dissipated in the load:
\begin{align}
P_{L} = V_T^2 \frac{R_L}{(R_T+R_L)^2}
\end{align}

(b) Show that the load is optimum when:
\begin{align*}
R_L = R_T
\end{align*}

Power in the load is max when $\partial P_L / \partial R_L = 0$
\begin{align}
\frac{\partial P_L }{ \partial R_L } & = V_T^2 \frac{(R_T+R_L)^2 - 2R_L(R_T+R_L)}{(R_T+R_L)^4} \\
& = V_T^2 \frac{R_T^2+2R_TR_L+R_L^2 - 2R_TR_L - 2R_L^2}{(R_T+R_L)^4} = \frac{R_T^2-R_L^2}{(R_T+R_L)^4}\\
&= \frac{(R_T-R_L)(R_T+R_L)}{(R_T+R_L)^4} = \frac{R_T-R_L}{(R_T+R_L)^3}
\end{align}
Thus,
\begin{align}
\frac{\partial P_L }{ \partial R_L } = 0 \quad \text{when}\quad R_L = R_T
\end{align}

\begin{align}
V^+ = V_{in(+)} \frac{R_2}{R_1+R_2} \quad \text(a)
\end{align}

2- Do the some on the bottom (inverting) branch:
\begin{align}
V^- - V_{out} & = (V_{in(-)} - V_{out} ) \frac{R_2}{R_1+R_2} \\
V^- & = V_{in(-)} \frac{R_2}{R_1+R_2} + V_{out} \frac{R_1}{R_1+R_2} \quad \text(b)
\end{align}

3- Ideal Op-Amp rule $V^+ = V^-$ and use eq. (a) and (b):
\begin{align}

### 4 Design of amplifier stages

One wishes to measure a pressure pulse with a very sensitive load cell. Unfortunately, the voltage output of the pressure transducer is $\pm 5$ mV and the smallest range of our Data Acquisition (DAQ) board is $\pm 1$ V with a desired bandwidth of 50 kHz.  The GBP of your amplifier stage is 1 MHz.

(a) To utilize the most of our DAQ board dynamic range, and hence increase our resolution, we wish to amplify our signal. What should be the gain to make the most of your DAQ system?

The pressure transducer signal should be amplified to 60-80% of the input of the DAQ system to prevent clipping.  We select a desired voltage of $V_{out} = 4\, \text{V}$ and $V_{in} = 5 \, \text{mV}$:

The nominal total gain, $G_{tot} = \frac{V_{out}}{V_{in}} = \frac{4}{0.005} = 800$


(b) How many amplifier stages would you need to successfully amplify your signal?

We have to follow two rules:

1- For the op-amp to operate in its ideal behavior and prevent impedance loading, the gain of each amplifier stage should not exceed $G_{stage} = 30$

2- We should also be concerned by the Gain Bandwidth Product, $GBP$ of the amplifier stages to make sure we have adequate bandwidth.

Here the desired bandwidth is $B = 50,000\,\text{Hz}$.  Since for our op-amps: $GBP = 10^6 \,\text{Hz}$, this forces a maximum gain per amplifier stages of:
\begin{align}
G_{stage} = \frac{GBP}{B} = \frac{10^6}{5\times10^4} = 20
\end{align}

We take the smaller values for the two conditions above and therefore the gain of each stage should not exceed $G_{stage} = 20$.  This condition is accomplished for three amplifier stages:
\begin{align}
G_{stage} = \sqrt[3]{G_{tot}} \approx 9.28
\end{align}
In practice, I will choose a gain of $G_{stage} = 9$, which will give me an overall gain of $G_{tot} = 729$ and an output voltage of $3.65\,\text{V}$, which is within my 60-80% input range of the DAQ system.
